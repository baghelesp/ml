```python
import numpy as np
import pandas as pd
data = pd.read_csv("/content/gdrive/MyDrive/ml_lab/play_tennis.csv")
data
```

```python
data = data[['outlook', 'temp', 'humidity', 'wind', 'play']]
data
```

```python
outlook = list(data['outlook'])
temp = list(data['temp'])
humidity = list(data['humidity'])
wind = list(data['wind'])
play = list(data['play'])
print(list(data['outlook']), len(list(data['outlook'])))
print(list(data['temp']), len(list(data['temp'])))
print(list(data['humidity']), len(list(data['humidity'])))
print(list(data['wind']), len(list(data['wind'])))
print(list(data['play']), len(list(data['play'])))
```

```python
#probability of yes & no
def probability(word):
  return play.count(word)/len(play)

#P(x1|Y)
def prob(a,b):
  c = 0
  if a in outlook:
    for i in range(len(play)):
      if outlook[i] == a and play[i] == b:
        c+=1
    return c
  elif a in temp:
    for i in range(len(play)):
      if temp[i] == a and play[i] == b:
        c+=1
    return c
  elif a in humidity:
    for i in range(len(play)):
      if humidity[i] == a and play[i] == b:
        c+=1
    return c
  elif a in wind:
    for i in range(len(play)):
      if wind[i] == a and play[i] == b:
        c+=1
    return c

#Driver code

#sunny, cool, high, strong
x1 = "Sunny"
x2 = "Cool"
x3 = "High"
x4 = "Strong"

#P(Y|x) = P(x1|Y) * P(x2|Y) * P(x3|Y) * P(x4|Y) * P(Y)
yes = (prob(x1, "Yes")/play.count("Yes")) * (prob(x2, "Yes")/play.count("Yes")) * (prob(x3, "Yes")/play.count("Yes")) * (prob(x4, "Yes")/play.count("Yes")) * probability("Yes")
no = (prob(x1, "No")/play.count("No")) * (prob(x2, "No")/play.count("No")) * (prob(x3, "No")/play.count("No")) * (prob(x4, "No")/play.count("No")) * probability("No")

if yes > no:
  print("RESULT OF NAIVE BAYES:")
  print(x1 +", "+ x2+ ", " + x3 + ", " + x4,"-> Tennis can be played in these conditions")
else:
  print("RESULT OF NAIVE BAYES:")
  print(x1 +", "+ x2+ ", " + x3 + ", " + x4,"-> Tennis cannot be played in these conditions")


print("\nProbability of playing tennis is: ", yes)
print("Probability of not playing tennis is: ", no)
```
